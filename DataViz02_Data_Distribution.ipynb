## 1.เขียน function ที่รับ input เป็น output ของ boxplot แล้ว แสดงค่า min, max, q1, q2, q3

```python
def box_vals(box_out_dict):
    q1 = ybp['boxes'][0].get_ydata()[0]
    q2 = ybp['medians'][0].get_ydata()[0]
    q3 = ybp['boxes'][0].get_ydata()[2]
    vmin =  min(ybp['whiskers'][0].get_xdata())
    vmax =  max(ybp['whiskers'][1].get_xdata())
    return print(f'q1={q1}, q2={q2}, q3={q3}, min={vmin}, max={vmax}')
```

```python
box_vals(df)  # ค่า q1,q2,q3,min,max
```
