# import packages ที่ต้องใช้ และนำไฟล์เข้ามาใน colab

```python
import pandas as pd  # import package เข้ามา
import os  # ใช้จัดการไฟล์ต่างๆ
```

```python
from google.colab import drive  # เชื่อม google drive เพื่อเอาข้อมูลมาใช้
drive.mount('/content/drive')  
path = '/content/drive/My Drive/ovrdt data'  # ชี้โฟลเดอร์ใน google drive
```

## ข้อมูลที่ 1 alldb

```python
alldb = pd.read_csv(os.path.join(path,'alldb.csv'))
alldb
```

```python
# เลือกเฉพาะ column ที่ต้องการ
alldb_data = alldb[['user',	'RowKey', 'date', 'phone', 'site', 'time', 'urine', 'ov', 'PartitionKey']]
alldb_data
```

## ข้อมูลที่ 2 hospital_list 

```python
hospital_list = pd.read_excel(os.path.join(path,'hospital_list.xlsx'))
hospital_list
```

```python
# เลือกเฉพาะ column ที่ต้องการ
hospital_data = hospital_list[['hcode',	'name', 'zone_name', 'province', 'amphur', 'tambon']]
hospital_data
```

# เชื่อมตาราง

```python
# เชื่อมตาราง alldb_data และ hospital_data เข้าด้วยกัน โดยให้ตาราง alldb_data เป็นตารางหลักอยู่ที่ด้านซ้าย
merge_table = alldb_data.merge(hospital_data,how='left',left_on='site',right_on='hcode')
merge_table
```

```python
# ตรวจสอบค่า missing ของตาราง merge_table
merge_table.isnull().any()
```

```python
# ทำการลบข้อมูลที่เป็นค่า missing ออก
merge_table_drop = merge_table.dropna()
merge_table_drop
```

```python
# ตรวจสอบค่า missing ของตาราง merge_table_drop
merge_table_drop.isnull().any()
```

```python
# จากการทำ dropna() ทำให้ข้อมูลหายไปกี่ % 
removed = merge_table.shape[0] - merge_table_drop.shape[0]

print(f'size before drop = {merge_table.shape[0]}')
print(f'size aftre drop = {merge_table_drop.shape[0]}')
print(f'we loss {100*(removed/merge_table.shape[0])}% of data')
```
