```python
import pandas as pd  # import pandas เพื่อมาจัดการตาราง
```

```python
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')  # ข้อมูลของดอกไม้ Iris
df
```

```python
df['Name'].unique()  # .unique ใช้เลือกเฉพาะค่าที่ไม่ซ้ำ
```

image.png

# Parallel Coordinates
https://pandas.pydata.org/docs/reference/api/pandas.plotting.parallel_coordinates.html

```python
pd.plotting.parallel_coordinates(df,'Name')  # Name ใช้แบ่งแยกสีแต่ละพันธุ์
# เส้นที่ plot ลงในกราฟเปรียบขนาดข้อมูล ไม่ใช่ข้อมูลอันไหนเยอะกว่า
```

### select column

```python
pd.plotting.parallel_coordinates(df,'Name',cols=['SepalLength','PetalLength']);  # cols= ใช้เลือกดูข้อมูลคอลัมน์
```

### change color

https://matplotlib.org/stable/tutorials/colors/colors.html

```python
pd.plotting.parallel_coordinates(df,'Name',color=['r','y','k']);  # เปลี่ยนสีข้อมูลในกราฟ
```

```python
pd.plotting.parallel_coordinates(df,'Name',color=[[0.5,0.5,0.5],[0,0,0.8],[0.2,0.4,0.7]]);  # เปลี่ยนสีข้อมูลในกราฟ โดยใช้การผสมสี
```

```python
pd.plotting.parallel_coordinates(df,'Name',color=['#FF00E0','#00FFE2','#ECFF00']);  # เปลี่ยนสีข้อมูลในกราฟ โดยใช้รหัสสี HTML
```

### Specify color by colormap

https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
pd.plotting.parallel_coordinates(df,'Name',colormap='Set3');  # เปลี่ยนสีข้อมูลในกราฟ โดยใช้เซตสีที่มีอยู่แล้ว
```

### Heatmap matrix

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html

```python
matplotlib.rcParams['figure.figsize']=[20,20]  # กำหนดขนาดของ figure
plt.imshow(df.iloc[:,:-1],interpolation='nearest')  # ไม่เอาคอลัมน์สุดท้ายมา plot
```

```python
plt.imshow(df.iloc[:,:-1],interpolation='nearest',cmap='hot')  # สีสว่างค่าน้อย สีเข้มค่ามาก
```

### Subplot

```python
matplotlib.rcParams['figure.figsize']=[10,10]  # กำหนดขนาดของ figure
```

```python
plt.subplot(1,3,1)  # subplot (มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,2)  # subplot (มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,3)  # subplot (มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[100:150,:-1],interpolation='nearest',cmap='hot')
```

```python
plt.subplot(1,2,1)  # subplot (มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,2,2)  # subplot (มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',cmap='hot',alpha=0.5)  # alpha กำหนดความโปร่งใสของสี 
```

```python
plt.subplot(1,4,1)  # subplot (มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,4,2)  # subplot (มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',cmap='hot',alpha=0.2)  # alpha กำหนดความโปร่งใสของสี 

plt.subplot(1,4,3)  # subplot (มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',cmap='hot',alpha=0.5)  # alpha กำหนดความโปร่งใสของสี 

plt.subplot(1,4,4)  # subplot (มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',cmap='hot',alpha=0.8)  # alpha กำหนดความโปร่งใสของสี 
```
