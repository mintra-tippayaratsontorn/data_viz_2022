# Parallel Coordinates
https://pandas.pydata.org/docs/reference/api/pandas.plotting.parallel_coordinates.html

```python
pd.plotting.parallel_coordinates(df,'Name')  # Name ใช้แบ่งแยกสีแต่ละพันธุ์
# เส้นที่ plot ลงในกราฟเปรียบขนาดข้อมูล ไม่ใช่ข้อมูลอันไหนเยอะกว่า
```

## HW finalGP 1

วาดกราฟ parallel coordinate เปรียบเทียบ [อายุ, เพศ] ของผู้ป่วย โดยใช้ class เป็น risk (เลือกมาแค่ 3 risks)

```python
import os  # จัดการไฟล์ต่างๆ
```

```python
from google.colab import drive  # เชื่อม google drive เพื่อเอาข้อมูลมาใช้
drive.mount('/content/drive')  # เชื่อมไปยังที่อยู่ google drive ที่สร้าง 
path = '/content/drive/My Drive/dataviz_2022_data'  # ชี้โฟลเดอร์ใน google drive
```

```python
data_file_path = os.path.join(path,'pm-20-jan-2021.csv')  # os.path ที่อยู่ไฟล์นี้ .join ชื่อไฟล์
data_df = pd.read_csv(data_file_path)  # pandas อ่าน csv
data_df
```

```python
data_hw = data_df[['age','sex','risk']]  # เลือกเฉพาะคอลัมน์ที่ต้องการคือ อายุ, เพศ, risk
data_hw
```

```python
data_hw.isnull().any()  # เช็ค missing ของตาราง data_hw
```

```python
hw_drop = data_hw.dropna()  # ทำการ drop ข้อมูลที่ missing
hw_drop
```

```python
hw_drop.isnull().any()  # เช็ค missing ของตาราง data_drop
```

```python
hw_drop['risk'].unique()  # เลือกดูข้อมูลคอลัมน์ risk ที่ไม่ซ้ำกัน
```

### เลือก risk ที่ต้องการ

```python
risk1 = hw_drop[hw_drop['risk'] == 'State Quarantine']  # เลือกข้อมูล risk ที่มาจากState Quarantine
risk1
```

```python
risk2 = hw_drop[hw_drop['risk'] == 'พิธีกรรมทางศาสนา']  # เลือกข้อมูล risk ที่มาจากพิธีกรรมทางศาสนา
risk2
```

```python
risk3 = hw_drop[hw_drop['risk'] == 'สนามมวย']  # เลือกข้อมูล risk ที่มาจากสนามมวย
risk3
```

### ต่อตารางใช้ concat (ต่อในแนวแกน Y)

```python
hw_concat = pd.concat([risk1,risk2,risk3])  # ต่อตารางในแนวแกน y
hw_concat
```

### แทนที่ข้อมูล

เปลี่ยน sex เป็นตัวเลขดังนี้
* ชาย เป็นเลข 0
* หญิง เป็นเลข 1

```python
hw_replace0 = hw_concat.replace('ชาย',0)  # แทนที่ ชาย ให้เป็นเลข 0
hw_replace0
```

```python
hw_replace1 = hw_replace0.replace('หญิง',1)  # แทนที่ หญิง ให้เป็นเลข 1
hw_replace1
```

 เปลี่ยนชื่อ risk ที่เลือกดังนี้
* พิธีกรรมทางศาสนา  =>  religious ritual
* สัมผัสใกล้ชิดกับผู้ป่วยยืนยันรายก่อนหน้านี้  =>  boxing stadium

```python
hw_replace2 = hw_replace1.replace('พิธีกรรมทางศาสนา','religious ritual')  # แทนที่ พิธีกรรมทางศาสนา ให้เป็น religious ritual
hw_replace2
```

```python
hw_replace3 = hw_replace2.replace('สนามมวย','boxing stadium')  # แทนที่ สนามมวย ให้เป็น boxing stadium
hw_replace3
```

### plot กราฟ

```python
pd.plotting.parallel_coordinates(hw_replace3, 'risk', color=['#FFA500','#9A0EEA','#AAFF32']);
```
