# เชื่อมตาราง

```python
# เชื่อมตาราง data1 และ data2 เข้าด้วยกัน โดยให้ตาราง alldb_data เป็นตารางหลักอยู่ที่ด้านซ้าย
data_d1d2 = data1.merge(data2,how='left',left_on='site',right_on='hcode')
data_d1d2
```

# เตรียมนำข้อมูลออก
