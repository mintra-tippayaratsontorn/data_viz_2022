# Load data

```python
import pandas as pd  # import package เข้ามา
import os  # ใช้จัดการไฟล์ต่างๆ
```

```python
from google.colab import drive  # เชื่อม google drive เพื่อเอาข้อมูลมาใช้
drive.mount('/content/drive')  
path = '/content/drive/My Drive/ovrdt data'  # ชี้โฟลเดอร์ใน google drive
```

# Data 1 alldb

```python
data1 = pd.read_csv(os.path.join(path,'alldb.csv'))  # สร้างตัวแปร รับค่าข้อมูลที่อยู่ใน Google drive
data1  # แสดงข้อมูล
```

```python
data1 = data1.drop(columns=['title','age','Unnamed: 0','care','pid'])  # ทำการลบคอลัมน์ที่ไม่ต้องการออกไป
print(data1.shape)
data1.head()
```

```python
data1.isnull().sum()
```

## user เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
data1 = data1[(data1['user'].astype(str).str.isdigit()) & (data1['user'].str.len() == 5)]  # จากนั้นเลือกมาเฉพาะ คอลัมน์ user ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 ตัว
data1  # แสดงตาราง
```

```python
data1.isnull().sum()  # ดูค่าว่าง
```

## RowKey เป็น String

```python
data1['RowKey']=data1['RowKey'].astype('string')  # โดยการเปลี่ยน dtype เป็น  String
data1  # แสดงตาราง
```

```python
print(data1.dtypes)  # จึงพบว่า RowKey เป็น String
```

## date เป็น YYYY-MM-DD

```python
data1['date'] = pd.to_datetime(data1 ['date'],format='%Y/%m/%d')  # เปลี่ยนรูปแบบของเเละจัดชนิดของข้อมูล(วันเดือนปี)
data1  # แสดงตาราง
```

```python
data1['date'] = data1[['date']].fillna(method="ffill")  # เติมค่าว่างของวันที่ด้วย ค่าก่อนหน้า
data1
```

```python
print(data1.dtypes)
print('*'*50)
data1.isnull().sum()
```

## site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
data1 = data1[(data1['site'].astype(str).str.isdigit()) & (data1['site'].str.len() == 5)]  # เลือกมาเฉพาะ คอลัมน์ site ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 ตัว
data1  # แสดงตาราง
```

```python
data1.isnull().sum()  # ดูค่าว่าง
```

## time เป็น hh-mm-ss

```python
data1['time'] = pd.to_datetime(data1['time'],format='%H:%M%:%S',errors='ignore')  # เปลี่ยนชนิดข้อมูลและจัดรูปแบบข้อมูล(วินาที นาที ชั่วโมง)
data1  # แสดงตาราง
```

```python
data1['time'] = data1[['time']].fillna(method="ffill")  # เติมค่าว่างของเวลาด้วย ค่าก่อนหน้า
data1
```

```python
data1.isnull().sum()  # ดูค่าว่าง
```

## urine มีค่า String

```python
data1['urine'] = data1[['urine']].fillna(method="ffill")  # เติมค่าว่างของ urine ด้วย ค่าก่อนหน้า
data1
```

```python
print(data1.dtypes)
```

```python
data1.isnull().sum()  # ดูค่าว่าง
```

## ov เป็นตัวเลข มีค่าตั้งแต่ 0-4 (0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)

```python
set(data1['ov'])  # เรียกดู set ของ ov
```

```python
data1 = data1[data1['ov']!='ติดเชื้อ +2 ']  # ไม่เอา 'ติดเชื้อ +2 '
data1  # แสดงตาราง
```

จากนั้นจะเหลือข้อมูล 38492 ข้อมูลเมื่อ ลบ

```python
set(data1['ov'])  # ดู set ของข้อมูล ติดเชื้อ +2
```

```python
data1[['ov']] = data1[['ov']].fillna('ไม่ระบุ')  # แทน ไม่ระบุ ในค่าว่าง
data1  # แสดงตาราง
```

## phone เป็น String

```python
data1['phone']=data1['phone'].astype('string')  # โดยการเปลี่ยน dtype เป็น  String
data1  # แสดงตาราง
```

```python
print(data1.dtypes)  # จึงพบว่า phone เป็น String
```

## PartitionKey เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
data1 = data1[(data1['PartitionKey'].astype(str).str.isdigit()) & (data1['PartitionKey'].str.len() == 5)]  # จากนั้นเลือกมาเฉพาะ คอลัมน์ PartitionKey ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 ตัว
data1  # แสดงตาราง
```

```python
data1.isnull().sum()  # ดูค่าว่าง
```

# Data 2 hospital_list

```python
hospital_list = pd.read_excel(os.path.join(path,'hospital_list.xlsx'))  # สร้างตัวแปร รับค่าข้อมูลที่อยู่ใน Google drive
hospital_list  # แสดงข้อมูล
```

```python
data2 = hospital_list[['hcode',	'name', 'zone_name', 'province', 'amphur', 'tambon']]  # เลือกเฉพาะ column ที่ต้องการ
data2
```

```python
data2.isnull().sum()  # ดูค่าว่าง
```

# เชื่อมตาราง

```python
# เชื่อมตาราง alldb_data และ hospital_data เข้าด้วยกัน โดยให้ตาราง alldb_data เป็นตารางหลักอยู่ที่ด้านซ้าย
merge_table = data1.merge(data2,how='left',left_on='site',right_on='hcode')
merge_table
```

```python
merge_table.isnull().any()  # ดูค่าว่าง
```

```python
merge_table_drop = merge_table.dropna()  # ลบข้อมูลที่เป็นค่าว่างออก
merge_table_drop
```

```python
# จากการทำ dropna() ทำให้ข้อมูลหายไปกี่ % 
removed = merge_table.shape[0] - merge_table_drop.shape[0]

print(f'size before drop = {merge_table.shape[0]}')
print(f'size aftre drop = {merge_table_drop.shape[0]}')
print(f'we loss {100*(removed/merge_table.shape[0])}% of data')
```

```python
merge_table_drop.isnull().any()  # ดูค่าว่าง
```
