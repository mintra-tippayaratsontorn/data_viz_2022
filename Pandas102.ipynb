## ต่อแกน X

```python
persanal_data = data_df[['No.','sex','age','nationality']]
persanal_data.head()
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']].sample(frac=1)
covid_data.head()
```

### เอาตารางสองอันมาต่อกัน

```python
persanalcovid_data = persanal_data.merge(covid_data)
persanalcovid_data.head()
```

```python
persanalcovid_data = persanal_data.merge(persanal_data)
persanalcovid_data.head()
```

# Groupby

https://www.kaggle.com/crawford/python-groupby-tutorial

```python
data_df.groupby('nationality').count()  # จัดกลุ่มข้อมูลโดยเลือกมา 1 column และนับข้อมูล
```

```python
data_df.groupby('nationality').mean()  # จัดกลุ่มข้อมูลโดยเลือกมา 1 column และหาค่าเฉลี่ย
```

```python
data_df.groupby('nationality')['age'].min()  # จัดกลุ่มข้อมูลโดยเลือกมา 1 column และต้องใส่ชื่อ column ถึงจะหาค่าน้อยที่สุดได้
```

```python
data_df.groupby('nationality').min(' ')  # จัดกลุ่มข้อมูลโดยเลือกมา 1 column และหาค่าน้อยที่สุดโดยทำแบบ .min(' ') แต่ถ้าทำแบบ .min() หาไม่ได้
```
